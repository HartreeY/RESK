```julia
function myfit(x, y)
    n = length(x)
    x_mean = mean(x)
    y_mean = mean(y)
    slope = sum((x .- x_mean) .* (y .- y_mean)) / sum((x .- x_mean) .^ 2)
    intercept = y_mean - slope * x_mean
    return slope, intercept
end
```
