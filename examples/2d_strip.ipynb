### 2d_strip.ipynb
#### Examples of axial range expansions in 2D.
---
First, include RESK:

```julia
include("../resk.jl")
include("../reskplots.jl")
```

Let's run a small simulation of an axial range expansion assuming the finite-sites model:

```julia
test = rangeexp_strip()
```

```julia
test["stats"]
```

```julia
re_heatmap(test,"pops")
```

Just as an example, let's run another simulation, this time specifying more generations, a smaller y-size for our world, a name, and that all data outputs should be returned. There's also a bottleneck after the burn-in region by default (can be turned off or changed via the `bottleneck` option).

```julia
test = rangeexp_strip(10,80;y_max=5,name="my test",data_to_generate="FPSN")
```

```julia
test["stats"]
```

Let's plot the evolution of deme-average fitness in time:

```julia
re_heatmap_fitn(test)
```

Let's also see what that deme-average heterozygosity distribution looks like (on a stringent logarithmic scale):

```julia
re_heatmap_Aasel(test;log_base=1.00001,clim=(0,50))
```

Since the (default) value of n_sel_loci (the number of selected loci out of n_loci=1000) in this simulation is 500, let's run a quick test to confirm that AA+Aa+aa=500:

```julia
re_heatmap(test["AAsel"]+test["Aasel"]+test["aasel"];clim=(0,640))
```

Let's get only the front values of deme-average fitness and plot them with the following:

```julia
test_fitn_frontarr = front_array(test,"fitn";oneside=true)
re_heatmap(test_fitn_frontarr,1,20)
```

And with the following, we can find the mean front fitness, and then plot its evolution in time:

```julia
test_fitn_frontav = average_front(test,"fitn")
plot(test_fitn_frontav,xlabel="Generation",label="Mean front fitness")
```

Likewise, we can find the population average:

```julia
test_fitn_av = average_all(test,"fitn")
plot(test_fitn_av,xlabel="Generation",label="Mean front fitness")
```
